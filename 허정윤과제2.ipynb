```python
# 과제 
# naver 뮤비 리뷰
# https://movie.naver.com/movie/point/af/list.naver
# 100페이지 -> 하나의 datafram -> csv파일로 저장
# 크롤링
# 번호 감상평 점수1(원래점수) 점수2(7점이상 1, 7점이하 0)
# 방법1. bs 이용
# 방법2. pandas.read_html() #html에서 table정보만 추출
# python 멀티프로세싱 모듈
```

```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup as bs
import matplotlib.pyplot as plt
import pandas as pd
from tqdm import tqdm
import csv
import time
```

```python
pages=list(range(1, 101))

for page in tqdm(pages):
    url = f'https://movie.naver.com/movie/point/af/list.naver?&page=' + str(page)
    html = requests.get(url)
    soup = bs(html.text,'html.parser')
    review = soup.find_all("td",{"class":"title"})

    review_data = []
    num=0
    for reviews in review:
        content = reviews.find("a",{"class":"report"}).get("onclick").split("', '")[2]
        if content != "":
            num +=1
            title = reviews.find("a",{"class":"movie color_b"}).get_text()
            score = reviews.find("em").get_text()
            if int(score) < 7:
                score2 = 0
            else:
                score2 = 1
                time.sleep(0.1)
            review_data.append([num, title,content,int(score),score2])
df_review = pd.DataFrame(review_data, columns=["번호", "제목", "내용", "점수", "점수2"])

with open ( "영화리뷰.csv", "w", newline ="",encoding = 'utf8' ) as f:
    write = csv.writer(f)
    write.writerow(df_review.columns)
    write.writerows(review_data)
```
